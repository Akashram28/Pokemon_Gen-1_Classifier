```python
poke_ids = {v:k for k,v in gen1_pokemon.items()}
```

```python
vid = cv2.VideoCapture(0)

pokemon_name = ""
while(True):
    ret, frame = vid.read()

    image = cv2.resize(frame,(80,80))
    image = np.asarray(image)
    image = np.array([image])
    # print(image.shape)
    predicted = model.predict(image)
    # print(np.argmax(predicted))
    pokemon_name = poke_ids[np.argmax(predicted)]

    frame = cv2.flip(frame, 1)
    frame = cv2.rectangle(frame,(0,0) , (300,60) ,(0,0,0),-1)
    frame = cv2.putText(frame, pokemon_name, (30,30), cv2.FONT_HERSHEY_SIMPLEX, 
                   0.5, (255, 0, 0), 2, cv2.LINE_AA)
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
  
# After the loop release the cap object
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
```
